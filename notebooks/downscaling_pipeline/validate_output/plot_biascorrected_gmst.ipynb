```python
%matplotlib inline 
import xarray as xr
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
direc = '/gcs/impactlab-data/climate/downscaling/qc/gmst'
filename = '{}_tasmax.nc'
```

```python
models = ['ACCESS-CM2', 'ACCESS-ESM1-5', 'AWI-CM-1-1-MR', 'FGOALS-g3', 'INM-CM4-8', 
          'INM-CM5-0', 'MIROC-ES2L', 'MIROC6']
for model in models:
    ds = xr.open_dataset(os.path.join(direc, filename.format(model)))
    fig, axes = plt.subplots(1, 1, figsize=(12, 4))
    for scenario in ds.scenario:
        ds['gmst'].sel(scenario=scenario).plot(label=scenario.values)
    '''    ds['gmst'].sel(scenario='ssp126').plot(label='ssp126')
        ds['gmst'].sel(scenario='ssp245').plot(label='ssp245')
        ds['gmst'].sel(scenario='ssp370').plot(label='ssp370')
        ds['gmst'].sel(scenario='ssp585').plot(label='ssp585')'''
    plt.legend()
    plt.title(model)
    plot_filepath = os.path.join(direc, 'figures', 'bias_corrected', '{}.png'.format(model))
    fig.savefig(plot_filepath, bbox_inches='tight')
    print("saved {} figure to {}".format(model, plot_filepath))
```
